```python
# Creation of Dictionary
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
}

print(thisdict)
```

```python
# Retriving value from Dictionary
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
    "year" : 2020
}

print(thisdict["brand"])
```

```python
# Overwriting the prexisting - dictionary
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
    "year" : 2020
}

print(thisdict)
```

```python
# To determine the Length of A Dictionary
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964,
    "year" : 2020
}

print(len(thisdict))
```

```python
# Changing/Adding Key : Value in Dictionary
car = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964
}

car["color"] = "red"
print(car)

x = car.values()
print(x)
```

```python
# To determine the type of the Variable
car = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1964
}

car["color"] = "red"
print(type(car))
```

```python
# Retrival of Values from a Dictionary
student = {
    "roll" : 101,
    "name" : "Amit",
    "Branch" : "ENTC",
}

student.pop("Branch")
print("After Removal : ", student)
```
